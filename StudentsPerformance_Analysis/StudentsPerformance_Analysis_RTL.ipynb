<div dir='rtl'>## المهمة 5: التحليل الإحصائي العام

نحسب المتوسط، الانحراف المعياري، وأعلى وأقل مجموع درجات.
</div>

```python
import numpy as np
average = np.mean(df['total_score'])
std_dev = np.std(df['total_score'])
max_score = np.max(df['total_score'])
min_score = np.min(df['total_score'])
print(f'متوسط المجموع: {average}')
print(f'الانحراف المعياري: {std_dev}')
print(f'أعلى مجموع: {max_score}')
print(f'أقل مجموع: {min_score}')
```

<div dir='rtl'>## المهمة 6: التصنيف والتجميع

نصنف الطلاب إلى ثلاث فئات حسب مجموع الدرجات:
- متميز
- عادي
- متعثر
</div>

```python
def classify(x):
    if x >= 240:
        return 'متميز'
    elif x < 150:
        return 'متعثر'
    else:
        return 'عادي'
df['category'] = df['total_score'].apply(classify)
f=df['category'].value_counts()
f.values
```

<div dir='rtl'>## المهمة 8: التصوير البياني

نرسم:
- هيستوجرام لتوزيع مجموع الدرجات
- مخطط أعمدة لعدد الطلاب في كل فئة
- Boxplot لمقارنة مجموع الدرجات بين الجنسين
</div>

```python
import matplotlib.pyplot as plt
# # رسم هيستوجرام يوضح توزيع مجموع درجات الطلاب
plt.hist(df['total_score'], bins=10, edgecolor='blue',color="yellow")  # تقسيم البيانات إلى 20 فئة مع تمييز حدود الأعمدة
plt.title('Distribution of Total Scores')  # عنوان الرسم البياني
plt.xlabel('Total Score')  # تسمية المحور الأفقي (مجموع الدرجات)
plt.ylabel('Number of Students')  # تسمية المحور الرأسي (عدد الطلاب في كل فئة)
plt.show()  # عرض الرسم البياني
#


# حساب عدد الطلاب في كل فئة تصنيفية
categories = df['category'].value_counts()

# رسم مخطط شريطي يوضح عدد الطلاب في كل فئة
plt.bar(categories.index, categories.values)
plt.title('Number of Students in Each Category')  # عنوان الرسم البياني
plt.xlabel('Category')  # تسمية المحور الأفقي (الفئة التصنيفية)
plt.ylabel('Number of Students')  # تسمية المحور الرأسي (عدد الطلاب)
plt.show()  # عرض الرسم البياني





```

```python
# رسم مخطط نقاط يوضح العلاقة بين درجات الرياضيات ودرجات القراءة لكل طالب
plt.scatter(df['math score'], df['reading score'])
plt.title('Math Score vs Reading Score')  # عنوان الرسم البياني
plt.xlabel('Math Score')  # تسمية المحور الأفقي (درجة الرياضيات)
plt.ylabel('Reading Score')  # تسمية المحور الرأسي (درجة القراءة)
plt.show()  # عرض الرسم البياني

# حساب عدد الطلاب في كل مستوى تعليم الوالدين
parental_education_counts = df['parental level of education'].value_counts()
print(parental_education_counts)
print(parental_education_counts.index)
print(parental_education_counts.values)

# رسم المخطط الشريطي الأفقي
plt.barh(parental_education_counts.index, parental_education_counts.values)
plt.title('Number of Students by Parental Education Level')  # عنوان الرسم البياني
plt.xlabel('Number of Students')  # تسمية المحور الأفقي (عدد الطلاب)
plt.ylabel('Parental Education Level')  # تسمية المحور الرأسي (مستوى تعليم الوالدين)
plt.show()  # عرض الرسم البياني


# رسم مخطط شريطي مكدس يوضح توزيع الطلاب حسب الجنس والفئة التصنيفية
pd.crosstab(df['gender'], df['category']).plot(kind='bar', stacked=True)
plt.title('Number of Students by Gender and Category')  # عنوان الرسم البياني
plt.xlabel('Gender')  # تسمية المحور الأفقي (الجنس)
plt.ylabel('Number of Students')  # تسمية المحور الرأسي (عدد الطلاب)
plt.show()  # عرض الرسم البياني


```

<div dir='rtl'>## المهمة 10: تمرين أو تطبيق إضافي للطالب

حساب متوسط كل مادة ورسم هيستوجرام لدرجات القراءة، وحساب عدد الطلاب الذين حضروا كورس التحضير وحققوا مجموع أعلى من المتوسط.
</div>

```python
mean_math = df['math score'].mean()
mean_reading = df['reading score'].mean()
mean_writing = df['writing score'].mean()
print("متوسط الرياضيات:", mean_math)
print("متوسط القراءة:", mean_reading)
print("متوسط الكتابة:", mean_writing)

# رسم هيستوجرام لتوزيع درجات القراءة
plt.hist(df['reading score'], bins=15, edgecolor='black')  # تقسيم البيانات إلى 15 فئة مع تمييز حدود الأعمدة باللون الأسود
plt.title('Reading Score Distribution')  # عنوان الرسم البياني
plt.xlabel('Reading Score')  # تسمية المحور الأفقي (درجة القراءة)
plt.ylabel('Number of Students')  # تسمية المحور الرأسي (عدد الطلاب)
plt.show()  # عرض الرسم البياني


above_avg = df[(df['test preparation course'] == 'completed') & (df['total_score'] > average)]
print("عدد الطلاب حضروا كورس التحضير وحققوا مجموع أعلى من المتوسط:", above_avg.shape[0])
```
