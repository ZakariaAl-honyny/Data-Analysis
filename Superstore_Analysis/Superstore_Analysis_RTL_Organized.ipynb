<div dir="rtl" align="center">

1- اقرأ ملف البيانات باستخدام pandas.csv_read

</div>


```python

# قراءة البيانات
SuperstoreData = pd.read_csv("SampleSuperstore.csv", encoding='cp1252')
SuperstoreData.head()
```

```python
# معلومات عن البيانات
SuperstoreData.info()
```

<div dir="rtl" align="center">

 2- تحقق من وجود قيم ناقصة، وقم بمعالجتها باتعويض أو حذف

</div>


```python

# القيم المفقودة
SuperstoreData.isnull().sum()
```

```python
# حذف الصفوف التي تحتوي على قيم فارغة
SuperstoreData.dropna(inplace=True)
# معلومات عن البيانات
SuperstoreData.info()
```

<div dir="rtl" align="center">

 3- تحقق من صحة البيانات مثل التواريخ والقيم الرقمية
</div>


```python

# تحويل التواريخ
SuperstoreData['Order Date'] = pd.to_datetime(SuperstoreData['Order Date'])
SuperstoreData['Ship Date'] = pd.to_datetime(SuperstoreData['Ship Date'])

# تحويل الارقام
SuperstoreData['Sales'] = pd.to_numeric(SuperstoreData['Sales'], errors='coerce')
SuperstoreData['Profit'] = pd.to_numeric(SuperstoreData['Profit'], errors='coerce')

# وصف البيانات
SuperstoreData.describe()
```

## 2 التحليل الأساسي

يشمل هذا الجزء حساب:
 - حساب إجمالي المبيعات (Sales) لكل ف ئة م نتج (Category)

 - حساب متوسط الربح (Profit) لكل م نطقة (Region)

 - تصنيف العملاء حسب حجم مشترياتهم بناًء على مجموع المبيعات يمكنك تقسيمهم إلى مجموعات حسب النسبة المئوية أو حدود تحددها

</div>

<div dir="rtl" align="center">

 1- حساب إجمالي المبيعات (Sales) لكل ف ئة م نتج (Category)

</div>


```python
SuperstoreData.groupby('Category')['Sales'].sum()
```

<div dir="rtl" align="center">

 2- حساب متوسط الربح (Profit) لكل م نطقة (Region)

</div>


```python
SuperstoreData.groupby('Region')['Profit'].mean()
```

<div dir="rtl" align="center">

 2- تصنيف العملاء حسب حجم مشترياتهم بناًء على مجموع المبيعات يمكنك تقسيمهم إلى مجموعات حسب النسبة المئوية أو حدود تحددها


</div>

```python
# إجمالي المبيعات لكل عميل وتصنيفهم
customer_sales = SuperstoreData.groupby('Customer Name')['Sales'].sum()

def classify_customer(sales):
    if sales <= 1000:
        return 'صغير'
    elif sales <= 5000:
        return 'متوسط'
    else:
        return 'كبير'

customer_category_alt2 = customer_sales.apply(classify_customer)
customer_category_alt2.value_counts()



# # تقسيم العملاء إلى فئات
# bins = [0, 1000, 5000, customer_sales.max()]
# labels = ['صغير', 'متوسط', 'كبير']  

# customer_category = pd.cut(customer_sales, bins=bins, labels=labels)
# customer_category.value_counts()
# # print(customer_category.value_counts())



# # ================= طريقة بديلة باستخدام qcut لتقسيم العملاء إلى ثلاث فئات متساوية الحجم
# customer_category_alt = pd.qcut(
#     customer_sales,
#     q=3,
#     labels=['صغير', 'متوسط', 'كبير']
# )

# customer_category_alt.value_counts()


```

## 4 الرسوم البيانية

المخططات المطلوبة:

1 مخطط شريطي (Chart Bar)لإظهار إجمالي المبيعات لكل فئة منتج

2 مخطط دائري (Chart Pie)لنسبة المبيعات لكل منطقة

3 مخطط نقاط (Plot Scatter) للعلاقة بين كمية المبيعات والربح

4 مخطط خطي (Chart Line) لعرض تطور المبيعات عبر الوقت إن ُوجد عمود تاريخ

5 أي رسم بياني آخر يمكن ك إنشاؤه يدعم تحليل


</div>


<div dir="rtl" align="center">

1- مخطط شريطي (Chart Bar) لإظهار إجمالي المبيعات لكل فئة منتج

</div>

```python
sales_by_category = SuperstoreData.groupby('Category')['Sales'].sum()
plt.figure()
sales_by_category.plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Number of Sales')
plt.show()
```

<div dir="rtl"  align="center">

2- مخطط دائري (Chart Pie) لنسبة المبيعات لكل منطقة

</div>

```python

region_sales = SuperstoreData.groupby('Region')['Sales'].sum()
plt.figure()
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Percentage by Region')
plt.show()
```

<div dir="rtl"  align="center">

3- مخطط نقاط (Plot Scatter) للعلاقة بين كمية المبيعات والربح

</div>

```python

plt.figure()
plt.scatter(SuperstoreData['Quantity'], SuperstoreData['Profit'])
plt.title('Relationship Between Quantity and Profit')
plt.xlabel('Quantity')
plt.ylabel('Profit')
plt.show()
```

<div dir="rtl"  align="center">

4- مخطط خطي (Chart Line) لعرض تطور المبيعات عبر الوقت إن ُوجد عمود تاريخ

</div>

```python

sales_over_time = SuperstoreData.groupby('Order Date')['Sales'].sum()
plt.figure()
plt.plot(sales_over_time)
plt.title('Sales Trend Over Time')
plt.show()
```

<div dir="rtl"  align="center">
